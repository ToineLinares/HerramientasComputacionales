## Variable number of arguments

```
# Variable number of arguments
def thetotal(*items):
    tot=0
    for i in items:
        tot+=i
    return tot
print(thetotal(1,2,10,-1))
print(thetotal(*[1,2,10,-1]))
```

## Recursive Functions

```
def factorial(n):
  if n == 0:
    return 1
  else: 
    return n*factorial(n-1)
```

```
factorial(6)
```

```
def fibonacci(n):
  if n == 1 or n == 2:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
```

```
fibonacci(6)
```

```
print([fibonacci(i) for i in range(20)])
```

## Exercise

Escriba una rutina que, usando una función recursiva definida por ustedes, sume
todos los números enteros de 1 a 100. La función debe imprimir 

*   *La suma de los enteros entre 1 y 100 es ’resultado’*

donde resultado es el resultado de la suma obtenido por su rutina.

$$\sum_1^n i  = \frac{(n)(n+1)}{2}$$

```
def sum_upto_integer(n):
  if n == 1:
    return 1
  else:
    return n + sum_upto_integer(n-1)
```

```
number = 300
print("La suma de los enteros entre 1 y {} es {}".format(number, sum_upto_integer(number)))
```
