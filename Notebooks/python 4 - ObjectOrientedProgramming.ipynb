## Inheritance

```
class Wolf: 
  def __init__(self, name, color):
    self.name = name
    self.color = color

  def bark(self):
    print("Grr...")

# Class Dog inherits the attributes and methods of Wolf
class Dog(Wolf):
  pass
  # def bark(self):
  #   print("Woof")
```

```
husky = Dog("Max", "grey")
husky.bark()
```

## Super



```
class Wolf: 
  def __init__(self, name, color):
    self.name = name
    self.color = color

  def bark(self):
    print("Grr...")

# Class Dog inherits the attributes and methods
class Dog(Wolf):
  #pass
   def bark(self):
     print("Woof")
     # Call the super class method
     super().bark()
```

```
husky = Dog("Max", "grey")
husky.bark()
```

## Example

One warrior and one knight with health = 50, the warrior has variable attack, the knight has an attack of 55.
The knight attacks first, check who dies first.

```
class Warrior:
  def __init__(self, attack):
    self.health = 50
    self.attack = attack
  def is_alive(self):
    if self.health <= 0:
      print('is dead')
    else:
      print('is alive')
```

```
class Knight(Warrior):
  def __init__(self):
    self.health = 50
    self.attack = 55
```

```
def fight(knight, warrior):
  knight.health = knight.health - warrior.attack
  print('Knight:')
  knight.is_alive()
  print('Warrior')
  warrior.is_alive()
  if knight.health > 0:
    warrior.health = warrior.health - knight.attack
    print('Knight:')
    knight.is_alive()
    print('Warrior')
    warrior.is_alive()
```

```
david = Warrior(35)
felipe = Knight()
```

```
fight(felipe, david)
```
